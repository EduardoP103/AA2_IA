**5. Calcular medidas estadísticas.**


*Podemos calcular medidas estadísticas para obtener una mejor comprensión de los datos.*
<br>
*Por ejemplo, podemos calcular el promedio y la desviación estándar del número de fallecidos en accidentes de tránsito.*



```python
import pandas as pd
import numpy as np

# Leer los datos del archivo CSV
df = pd.read_csv("/content/drive/MyDrive/Data Python/Base_datos_accidentes.csv", encoding="latin-1", sep=';')

# Eliminar la cadena "N.I." de la columna "FALLECIDOS"
df['FALLECIDOS'] = df['FALLECIDOS'].replace('N.I.', np.NaN)

# Convertir la columna "FALLECIDOS" a tipo numérico
df['FALLECIDOS'] = pd.to_numeric(df['FALLECIDOS'])

# Calcular el promedio y la desviación estándar de la columna "FALLECIDOS"
promedio_fallecidos = df['FALLECIDOS'].mean()
desviacion_estandar_fallecidos = df['FALLECIDOS'].std()

# Imprimir los resultados
print('Promedio de fallecidos:', promedio_fallecidos)
print('Desviación estándar de fallecidos:', desviacion_estandar_fallecidos)
```

 El promedio de fallecidos es de 0.1689 personas, con una desviación estándar de 0.7925 personas. Esto significa que la mayoría de los accidentes de tránsito no resultan en fatalidades, y que la mayoría de los accidentes que resultan en fatalidades involucran a una sola persona.

```python
porcentaje_fatalidades = df['FALLECIDOS'].sum() / df.shape[0] * 100
print('Porcentaje de accidentes con fatalidades:', porcentaje_fatalidades)
```

**6. Interpretar los resultados.**

El porcentaje de accidentes con fatalidades es de aproximadamente 16,9%. Esto significa que aproximadamente 1 de cada 6 accidentes de tránsito resulta en una fatalidad.
